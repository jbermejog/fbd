# Gráficas del Iris Dataset - Para el Pétalo

**Principales dependencias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**DataFrame**

```python
url="https://raw.githubusercontent.com/jbermejog/fbd/master/2_IrisSpecies.csv"
df = pd.read_csv(url)
df.head()
```

**Gráfico con pandas - pétalo**
Gráfica sépalo - longitud vs ancho

```python
figura = df[df.Species == 'Iris-setosa'].plot(kind='scatter',
                                         x = 'PetalLengthCm', y = 'PetalWidthCm',
                                         color='blue', label ='Setosa',
                                         figsize=(6.4,6.4))
df[df.Species == 'Iris-versicolor'].plot(kind='scatter',
                                         x = 'PetalLengthCm', y = 'PetalWidthCm',
                                         color='orange', label ='Versicolor',
                                         ax=figura)
df[df.Species == 'Iris-virginica'].plot(kind='scatter',
                                         x = 'PetalLengthCm', y = 'PetalWidthCm',
                                         color='green', label ='Virginica',
                                         ax=figura)
figura.set_xlabel('longitud')
figura.set_ylabel('Anchura')
figura.set_title('Gráfica para el Pétalo - con Pandas')
plt.show()
```

# Gráfico con MATPLOTLIB - Pétalo

```python
df_set = df[df.Species=='Iris-setosa']
df_vers = df[df.Species=='Iris-versicolor']
df_virg = df[df.Species=='Iris-virginica']

fig = plt.figure(figsize = (6.4,6.4))
ax1 = fig.add_subplot(111)
ax1.scatter(df_set["PetalLengthCm"], df_set["PetalWidthCm"], s=10, c='blue', label='Setosa')
ax1.scatter(df_vers["PetalLengthCm"], df_vers["PetalWidthCm"], s=10, c='orange', label='Versicolor')
ax1.scatter(df_virg["PetalLengthCm"], df_virg["PetalWidthCm"], s=10, c='green', label='Virginica')
plt.legend(loc='upper left')
plt.title('Gráfica para el pétalo - con Matplotlib')
plt.xlabel('Longitud')
plt.ylabel('Anchura')
plt.show()
```

Otro ejemplo utilizando un dataframe con localizaciones de estaciones de bicicleta de dublin para que se muestren en el mapa

```python
import pandas as pd
import folium

estaciones_dublin = "https://raw.githubusercontent.com/jbermejog/fbd/master/dublin.csv"
localizaciones = pd.read_csv(estaciones_dublin)
localizaciones.head()
```

Seleccionamos solo la posicion y en nombre

```python
localizaciones = localizaciones[["Latitude", "Longitude", "Name"]]
```

Pintamos el mapa centrado usando mean

```python
map = folium.Map(location=[localizaciones.Latitude.mean(), localizaciones.Longitude.mean()], zoom_start=14, control_scale=True)
map
```

Agregamos los puntos al mapa

```python
for index, location_info in localizaciones.iterrows():
    folium.Marker([location_info["Latitude"], location_info["Longitude"]], popup=location_info["Name"]).add_to(map)
map
```
